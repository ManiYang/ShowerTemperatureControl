```python
def read_file(filename):
    """
    Return (ts[], Ts[], T_corr[])
    """
    def parse_line(line):
        """
        Return (time, temperature).
        """
        line = line.strip()
    #     print(line)
        h = int(line[0:2])
        m = int(line[3:5])
        s = float(line[6:12])
        T = float(line[16:21])
        Tcorr = float(line[22:])
    #     print(h, m, s, T)
        return (h*60 + m)*60 + s, T, Tcorr

    ts = []
    Ts = []
    T_corr = []
    with open(filename, 'r') as f:
        for line in f:
            t, T, Tcorr = parse_line(line)
            ts.append(t)
            Ts.append(T)
            T_corr.append(Tcorr)
    ts = [t-ts[0] for t in ts]

    print('last T:', Ts[-1])
    return ts, Ts, T_corr
```

```python
import matplotlib.pyplot as plt

filename = 'test1.txt' # correct with criterion
ts, Ts, T_corr = read_file(filename)
T_last = Ts[-10]

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,1,1)

xlim = [10, 50]
ax.plot(xlim, [T_last,T_last], label='', color='#aaaaff')
ax.plot(xlim, [T_last-1,T_last-1], label='', color='#ccccff')
ax.plot(ts, Ts, '.', label='measure')
ax.plot(ts, T_corr, '.-', label='corrected')
ax.set(xlim=xlim, xlabel='t (s)', ylabel='T (deg C)')
# ax.set(ylim=[38,45])
plt.grid()
plt.legend(loc='best')
plt.show()
```
